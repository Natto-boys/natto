```python
# imports
import pygsheets

from sheets import SPREADSHEET_ID, get_sheet_as_df

# TODO refactor out of app.py and import
OPTIONS = ["Funny", "Unsure", "Not Funny"]

gc = pygsheets.authorize("client_secret.json")
```

```python
# enter sheetname
SHEETNAME = "6308e06"
```

```python
# load df from google sheet
sh = gc.open_by_key(SPREADSHEET_ID)
df = get_sheet_as_df(sh, SHEETNAME)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# plot with label values on x axis, count of each value on y axis (histogram), with counts split by recipe.
# make sure x axis is sorted in same order as OPTIONS
sns.set_theme(style="darkgrid")
ax = sns.countplot(x="label", hue="recipe", data=df, order=OPTIONS)
plt.show()
```
